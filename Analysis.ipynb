```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = '/content/pl-tables-2004-2024.csv'
pl_data = pd.read_csv(file_path)

# Filter top and bottom club positions
top_positions = [1, 2, 3, 4]
bottom_positions = [17, 18, 19, 20]

top_clubs = pl_data[pl_data['position'].isin(top_positions)]
bottom_clubs = pl_data[pl_data['position'].isin(bottom_positions)]

# Calculate average points, goals for, and goals against per season for top and bottom clubs
top_clubs_avg = top_clubs.groupby('Year').agg({'points': 'mean', 'gf': 'mean', 'ga': 'mean'}).reset_index()
bottom_clubs_avg = bottom_clubs.groupby('Year').agg({'points': 'mean', 'gf': 'mean', 'ga': 'mean'}).reset_index()

# Display initial data for verification
print("Top Clubs Averages (First 5 Rows):")
print(top_clubs_avg.head())
print("\nBottom Clubs Averages (First 5 Rows):")
print(bottom_clubs_avg.head())

# Set up the figure and axes for plotting
plt.figure(figsize=(15, 10))

# Plot Average Points
plt.subplot(3, 1, 1)
plt.plot(top_clubs_avg['Year'], top_clubs_avg['points'], label='Top Clubs - Average Points', color='blue', linestyle='-', marker='o')
plt.plot(bottom_clubs_avg['Year'], bottom_clubs_avg['points'], label='Bottom Clubs - Average Points', color='red', linestyle='--', marker='x')
plt.ylabel('Average Points')
plt.title('Average Points per Season')
plt.legend()
plt.grid(True)
plt.grid(which='minor', linestyle=':', linewidth=0.5)  # Enable sub-grid

# Plot Average Goals For
plt.subplot(3, 1, 2)
plt.plot(top_clubs_avg['Year'], top_clubs_avg['gf'], label='Top Clubs - Goals Scored', color='blue', linestyle='-', marker='o')
plt.plot(bottom_clubs_avg['Year'], bottom_clubs_avg['gf'], label='Bottom Clubs - Goals Scored', color='red', linestyle='--', marker='x')
plt.ylabel('Average Goals Scored')
plt.title('Average Goals Scored per Season')
plt.grid(True)
plt.grid(which='minor', linestyle=':', linewidth=0.5)  # Enable sub-grid

# Plot Average Goals Against
plt.subplot(3, 1, 3)
plt.plot(top_clubs_avg['Year'], top_clubs_avg['ga'], label='Top Clubs - Goals Conceded', color='blue', linestyle='-', marker='o')
plt.plot(bottom_clubs_avg['Year'], bottom_clubs_avg['ga'], label='Bottom Clubs - Goals Conceded', color='red', linestyle='--', marker='x')
plt.ylabel('Average Goals Conceded')
plt.xlabel('Season End Year')
plt.title('Average Goals Conceded per Season')
plt.grid(True)
plt.grid(which='minor', linestyle=':', linewidth=0.5)  # Enable sub-grid

# Enable minor ticks for sub-grids on x-axis
for ax in plt.gcf().axes:
    ax.set_xticks(range(2004, 2025))
    ax.set_xticklabels(range(2004, 2025), rotation=45)
    ax.minorticks_on()  # Turn on minor ticks

# Adjust layout and show plot
plt.tight_layout(pad=2.0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = '/content/pl-tables-2004-2024.csv'
pl_data = pd.read_csv(file_path)

# Display the first few rows for verification
print(pl_data.head())

# Get team name input
team_name = input("Enter the team name: ")

# Filter data for the specified team
team_data = pl_data[pl_data['Team'].str.contains(team_name, na=False, case=False)]
if not team_data.empty:
    # Calculate average points, goals for, and goals against per season for the team
    team_data_avg = team_data.groupby('Year').agg({
        'points': 'mean',
        'gf': 'mean',
        'ga': 'mean'
    }).reset_index()

    # Set up the figure and axes for plotting
    plt.figure(figsize=(14, 10))

    # Plot Average Points
    plt.subplot(3, 1, 1)
    plt.plot(team_data_avg['Year'], team_data_avg['points'], label=f'{team_name} - Points', color='blue', linestyle='-', marker='o')
    plt.ylabel('Average Points')
    plt.title(f'{team_name} Performance: Average Points per Season')
    plt.legend()
    plt.grid(True)
    plt.grid(which='minor', linestyle=':', linewidth=0.5)  # Enable sub-grid

    # Plot Average Goals For
    plt.subplot(3, 1, 2)
    plt.plot(team_data_avg['Year'], team_data_avg['gf'], label=f'{team_name} - Goals Scored', color='green', linestyle='-', marker='o')
    plt.ylabel('Average Goals Scored')
    plt.title(f'{team_name} Performance: Average Goals Scored per Season')
    plt.grid(True)
    plt.grid(which='minor', linestyle=':', linewidth=0.5)  # Enable sub-grid

    # Plot Average Goals Against
    plt.subplot(3, 1, 3)
    plt.plot(team_data_avg['Year'], team_data_avg['ga'], label=f'{team_name} - Goals Conceded', color='red', linestyle='-', marker='o')
    plt.ylabel('Average Goals Conceded')
    plt.xlabel('Season End Year')
    plt.title(f'{team_name} Performance: Average Goals Conceded per Season')
    plt.grid(True)
    plt.grid(which='minor', linestyle=':', linewidth=0.5)  # Enable sub-grid

    # Set x-axis ticks to show each year and enable minor ticks
    for ax in plt.gcf().axes:
        ax.set_xticks(range(2004, 2024))  # Show each year
        ax.set_xticklabels(range(2004, 2024), rotation=45)  # Rotate for readability
        ax.minorticks_on()  # Turn on minor ticks for sub-grids

    # Adjust layout and show plot
    plt.tight_layout(pad=2.0)
    plt.show()

else:
    print(f"No data found for team '{team_name}'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = '/content/valuation.csv'
data = pd.read_csv(file_path)
print("Available columns:", data.columns)
print(data.head())

# Get team name from user
team_name = input("Enter the team name: ")
team_data = data[data['Team'].str.lower() == team_name.lower()]

# Filter data to start from 2004
team_data = team_data[team_data['Year'] >= 2004]

# Set up the figure with three stacked subplots
plt.figure(figsize=(10, 16))

# Plot Income vs Expenditure
plt.subplot(3, 1, 1)
plt.plot(team_data['Year'], team_data['Income'], label='Income', marker='o', color='blue')
plt.plot(team_data['Year'], team_data['Expenditure'], label='Expenditure', marker='o', color='red')
plt.title('Income vs Expenditure')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.xticks(range(2011, team_data['Year'].max() + 1))  # Show every year from 2004
plt.legend()
plt.grid(which='major', linestyle='-', linewidth=0.75, color='grey')  # Major grid lines
plt.grid(which='minor', linestyle=':', linewidth=0.5, color='lightgrey')  # Minor sub-grid lines
plt.minorticks_on()  # Enable minor ticks for sub-grid lines

# Plot Net Profit separately
plt.subplot(3, 1, 2)
net_profit = team_data['Income'] - team_data['Expenditure']
plt.plot(team_data['Year'], net_profit, label='Net Profit', marker='o', color='purple')
plt.title('Net Profit')
plt.xlabel('Year')
plt.ylabel('Net Profit')
plt.xticks(range(2011, team_data['Year'].max() + 1))  # Show every year from 2004
plt.legend()
plt.grid(which='major', linestyle='-', linewidth=0.75, color='grey')  # Major grid lines
plt.grid(which='minor', linestyle=':', linewidth=0.5, color='lightgrey')  # Minor sub-grid lines
plt.minorticks_on()  # Enable minor ticks for sub-grid lines

# Plot Valuation
plt.subplot(3, 1, 3)
plt.plot(team_data['Year'], team_data['Valuation'], label=f'{team_name} Valuation', marker='o', color='green')
plt.title(f'Valuation of {team_name}')
plt.xlabel('Year')
plt.ylabel('Valuation')
plt.xticks(range(2011, team_data['Year'].max() + 1))  # Show every year from 2004
plt.legend()
plt.grid(which='major', linestyle='-', linewidth=0.75, color='grey')  # Major grid lines
plt.grid(which='minor', linestyle=':', linewidth=0.5, color='lightgrey')  # Minor sub-grid lines
plt.minorticks_on()  # Enable minor ticks for sub-grid lines

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Load the dataset
file_path = '/content/valuation.csv'
df = pd.read_csv(file_path)

# Display the first few rows to understand the data structure
print("Data Overview:")
print(df.head())

# Ensure that the expected columns are in the DataFrame
required_columns = ['Team', 'Valuation', 'Points']
for col in required_columns:
    if col not in df.columns:
        raise ValueError(f"Expected column '{col}' not found in the dataset.")

# User input for the team name
team_name = input("Enter the team name for analysis: ")

# Filter the data for the selected team
team_data = df[df['Team'] == team_name]

# Check if the team exists in the data
if team_data.empty:
    print(f"Team '{team_name}' not found in the dataset.")
else:
    # Create a scatter plot with Points on the x-axis and Valuation on the y-axis
    plt.figure(figsize=(10, 8))
    plt.scatter(team_data['Points'], team_data['Valuation'], color='r', marker='o', label=f'{team_name} Valuation')

    # Connect the dots with a line
    plt.plot(team_data['Points'], team_data['Valuation'], color='r', linestyle='-', linewidth=1)

    # Adding a trendline using Linear Regression
    X = team_data['Points'].values.reshape(-1, 1)  # Reshape to 2D array
    y = team_data['Valuation'].values
    model = LinearRegression()
    model.fit(X, y)
    trendline = model.predict(X)

    # Plotting the trendline
    plt.plot(team_data['Points'], trendline, color='b', linestyle='--', label='Trendline')

    # Title and labels
    plt.title(f"Valuation vs. Performance for {team_name}")
    plt.xlabel('Points')
    plt.ylabel('Valuation (in Millions)')

    # Annotating the highest point on the trendline
    highest_point_index = team_data['Valuation'].idxmax()
    plt.annotate(f"Highest Valuation\n({team_data['Points'][highest_point_index]}, {team_data['Valuation'][highest_point_index]})",
                 (team_data['Points'][highest_point_index], team_data['Valuation'][highest_point_index]),
                 textcoords="offset points", xytext=(0, 10), ha='center', fontsize=10, color='black')

    # Customize grid lines to show sub-grids as well
    plt.grid(which='major', linestyle='-', linewidth=0.75, color='grey')  # Major grid lines
    plt.grid(which='minor', linestyle=':', linewidth=0.5, color='lightgrey')  # Minor sub-grid lines
    plt.minorticks_on()  # Enable minor ticks for sub-grid lines

    # Display the legend and show the plot
    plt.legend()
    plt.show()
```

```python
import pandas as pd

# Load the datasets
performance_data = pd.read_csv('/content/pl-tables-2004-2024.csv')
valuation_data = pd.read_csv('/content/valuation.csv')

# Merge datasets on 'Season' and 'Club'
data = pd.merge(performance_data, valuation_data, how='inner', on=['Year', 'Team'])

# Filter relevant columns
data = data[['Year', 'Team', 'position', 'Valuation']]

# Convert Position to numeric if needed and Valuation should already be numeric
data['position'] = pd.to_numeric(data['position'], errors='coerce')

# Group by club and calculate mean and standard deviation of valuation and position
club_stability = data.groupby('Team').agg(
    avg_valuation=('Valuation', 'mean'),
    valuation_std=('Valuation', 'std'),
    avg_position=('position', 'mean'),
    position_std=('position', 'std')
).reset_index()

# Sort clubs by high average valuation and low valuation_std for consistency
consistent_high_valuation_clubs = club_stability.sort_values(by=['avg_valuation', 'valuation_std'], ascending=[False, True])

# Sort clubs by low average position (higher position) and low position_std for performance consistency
consistent_performance_clubs = club_stability.sort_values(by=['avg_position', 'position_std'], ascending=[True, True])

print("Clubs with consistent high valuations:")
print(consistent_high_valuation_clubs[['Team', 'avg_valuation', 'valuation_std']])

print("\nClubs with consistent high performance:")
print(consistent_performance_clubs[['Team', 'avg_position', 'position_std']])
```
